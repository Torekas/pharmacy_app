```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
file_path = 'check_status_stats.csv'
file_path = '10k_search_stats.csv'
# file_path = '10k_add_to_cart.csv'
df = pd.read_csv(file_path)

# Plotting the data
plt.figure(figsize=(14, 8))

plt.plot(df['tps'], label='Transactions Per Second (TPS)')
plt.plot(df['min'], label='Min response time (ms)')
plt.plot(df['max'], label='Max response time (ms)')
plt.plot(df['avg'], label='Average response time (ms)')
plt.plot(df['last'], label='Last response time (ms)')

# Adding titles and labels
plt.title('''Load test - search by name
GET http://localhost:5001/search?product_name=Aspirin
9999 processes, 1000ms delay, 30s duration''')
# plt.yscale('log')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend(loc='lower right')
plt.grid(True)

# Show plot
plt.show()
```
